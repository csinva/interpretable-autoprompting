## Compute results

For each task and n_shots, what is the rank of the first value containing that task's provided keyword?

```python
import sys
sys.path.append('..')
```

```python
from typing import Callable

import re

from data import TASKS_ONE_NUM, TASKS_TWO_NUMS, TASKS_ANLI
TASKS = {**TASKS_TWO_NUMS, **TASKS_ONE_NUM, **TASKS_ANLI}

def get_check_answer_func_for_task(task_name: str) -> Callable[str, bool]:
    check_answer_func = TASKS[task_name]['check_answer_func']
    if isinstance(check_answer_func, str):
        check_answer_func_re = re.compile(
            check_answer_func, re.IGNORECASE).search
    def check_answer_func(x): return bool(check_answer_func_re(x))
    return check_answer_func
```

```python
for task_name in df['task_name'].unique():
    check_answer_func = get_check_answer_func_for_task(task_name=task_name)
    df[f'{task_name}_check_answer_result'] = df['prefixes'].map(check_answer_func)
```

```python
ranks = []

for task_name in df['task_name'].unique():
    for n_shots in df['n_shots'].unique():
        df_subset = df[(df['task_name'] == task_name) & (df['n_shots'] == n_shots)]
        # print(df_subset.head())
        df_subset = df_subset.sort_values(by='accs', ascending=False).reset_index()
        df_subset = df_subset[df_subset[f'{task_name}_check_answer_result']]
        # print(df_subset.head())
        index_of_first_correct_prompt = df_subset.index.min()
        print(index_of_first_correct_prompt, '\t', task_name, n_shots)
        ranks.append((index_of_first_correct_prompt, task_name, n_shots))
    print('*' * 40 + '\n')
```

```python
ranks_df = pd.DataFrame(ranks, columns=['rank', 'task_name', 'n_shots'])
```

```python
import seaborn as sns
sns.set_style('white')
```

```python
ranks_df_pivot = ranks_df.pivot(index='n_shots', columns='task_name', values='rank')
ranks_df_pivot.head()
```

```python
from typing import Dict


def get_color_for_row(row: Dict) -> str:
    task_name = row['task_name']
    was_correct = row[f'{task_name}_check_answer_result']
    return red if was_correct else blue
    
df['hue'] = df.apply(get_color_for_row, axis=1)
```

```python
import matplotlib.pyplot as plt

f, axes = plt.subplots(2, 2, figsize=(20, 16))
sns.set(font_scale=1.5)

for idx in range(4):
    n_shots = 2**(idx)
    print(n_shots)
    oneshot_df = df[df['n_shots'] == n_shots].reset_index()

    r = idx // 2
    c = idx % 2
    ax = axes[r][c]
    sns.stripplot(
        data=oneshot_df, y='task_name', x='accs', hue='hue', size=10, alpha=0.8, marker="D", ax=ax
    )
    ax.set_xlim((0.0, 1.0))
    ax.get_legend().remove() # remove legend
    ax.set_title(f'{n_shots}-shot accuracy across various prompts')
    ax.set_xlabel('Accuracy')
    ax.set_ylabel(None)
```
